El dataset conté 98 files i 21 columnes amb les variables següents:

- Rk: Rang (classificació) del club dins la UEFA.
- Squad: Nom del club de football.
- Country: País al qual pertany el club.
- LgRk: Classificació del club dins la seva lliga.
- MP: Nombre total de partits jugats pel club.
- W: Nombre de partits guanyats pel club.
- D: Nombre de partits empatats pel club.
- L: Nombre de partits perduts pel club.
- GF: Nombre de gols a favor (gols marcats pel club).
- GA: Nombre de gols en contra (gols rebut per club).
- GD: Diferència de gols (GF - GA).
- Pts: Puntuació total obtinguda pel club en la lliga.
- Pts/G: Puntuació mitjana per partit jugat.
- xG: Expectativa de gols per al club.
- xGA: Expectativa de gols en contra per al club.
- xGD: Diferència esperada de gols (xG - xGA).
- xGD/90: Diferència esperada de gols per 90 minuts de joc.
- Last 5: Rendiment del club en els últims 5 partits.
- Attendance: Assistència mitjana als partits del club.
- Top Team Scorer: Màxim golejador del club.
- Goalkeeper: Porter principal del club.

```python
#Relació entre variables
import matplotlib.pyplot as plt
import seaborn as sns

variables_numeriques = ['Rk', 'LgRk', 'MP', 'W', 'D', 'L', 'GF', 'GA', 'GD',
       'Pts', 'Pts/G', 'xG', 'xGA', 'xGD', 'xGD/90', 'Attendance']

corrmat = UEFA_data[variables_numeriques].corr()
f, ax = plt.subplots(figsize=(12, 9))
sns.heatmap(corrmat, annot=True, vmax=.8, square=True, cmap='coolwarm');
plt.show()
```

En general les variables tenen una forta correlació entre sí. Les variables amb una relació més forta són Pts i Pts/G, de 0.99. Això té sentit en la mesura que, com més augmenta la puntuació mitjana per partit jugat, més augmenta la puntuació total obtinguda pel club en la lliga.

Aquestes dues variables també tenen una correlació de 0.99 amb la variable W, relativa al nombre de partits guanyats pel club. Això s'explica perquè en cas de victòria l'equip obté una major puntuació, de manera que a més victòries, més punts acumularà, tant per partit com en la classificació.

D'altra banda, la variable Rk té una forta correlació negativa amb les variables Pts i Pts/G, de -0.98. El motiu és que com més punts acumula l'equip, més alt escala en la classificació. 

També destacar que les variables xGD (Diferència esperada de gols) i xGD/90 (Diferència esperada de gols per 90 minuts de joc) tenen una correlació total perquè els partits són de 90 minuts.

#### Exercici 1.  Selecciona un atribut del conjunt de dades. Calcula el p-valor i digues si rebutja la hipòtesi nul·la agafant un alfa de 5%. 

```python
from scipy.stats import shapiro 
data = UEFA_data['W']
stat, p = shapiro(data)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
    print('Probably Gaussian')
else:
    print('Probably not Gaussian')
```

El p-value és 0.011, inferior a alpha (0.05). Per tant, es rebutja la hipòtesis nul·la, fet que indica que hi ha una evidència signficativa per concloure que la mostra no segueix una distribució normal.

```python
import seaborn as sns
sns.histplot(UEFA_data, x='W')
```

Tal i com podem veure en aquest histograma, la variable W, relativa al nombre de partits guanyats pel club, no segueix una distribució normal. Veiem que la distribució no és simètrica i està esbiaixada cap a la dreta. Això vol dir que tot i que la majoria de clubs acumulen uns 13 partits guanyats, hi ha alguns equips que destaquen per haver guanyat molts més partits. 

Aquesta distribució asimètrica es veu reflexada en la mitjana, que és superior a la mediana, amb quasi 0,5 de diferència.

```python
print('Mitjana:', UEFA_data['W'].mean())
print('Mediana:', UEFA_data['W'].median())
```

#### Exercici 2. Amb el mateix dataset selecciona dos altres atributs del conjunt de dades. Calcula els p-valors i digues si rebutgen la hipòtesi nul·la agafant un alfa de 5%.

```python
from scipy.stats import pearsonr
data1 = UEFA_data['GF']
data2 = UEFA_data['Pts']
stat, p = pearsonr(data1, data2)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
    print('Probably independent')
else:
    print('Probably dependent')
```

El p-value és 0.00, inferior a alpha (0.05). Per tant, es rebutja la hipòtesis nul·la, fet que indica que hi ha una evidència signficativa per concloure que hi ha una dependència entre les dues mostres.

```python
sns.scatterplot(UEFA_data, x='GF', y='Pts')
```

La gràfica mostra que hi ha una relació lineal forta entre les dos variables. Això vol dir que com més gols a favor acumula un equip, major és la seva puntuació en la lliga. 

L'anterior també ens permet deduïr que un major nombre de gols està associat a un major nombre de victòries i, per tant, a un major nombre de punts per partit i en la classifcació.

#### Exercici 3. Continua amb el conjunt de dades adjunt i selecciona tres atributs del conjunt de dades. Calcula el p-valor i digues si rebutja la hipòtesi nul·la agafant un alfa de 5%.

```python
from scipy.stats import kruskal
data1 = UEFA_data['xG']
data2 = UEFA_data['xGA']
data3 = UEFA_data['xGD']
stat, p = kruskal(data1, data2, data3)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
    print('Probably the same distribution')
else:
    print('Probably different distributions')
```

El p-value és 0.000, inferior a alpha (0.05). Per tant, es rebutja la hipòtesis nul·la, fet que indica que hi ha una evidència significativa per concloure que les mostres tenen distribucions diferents.

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(10,8))

ax1 = fig.add_subplot(1, 3, 1)

ax1.hist(UEFA_data['xG'])
plt.title('Expectativa de gols')

ax2 = fig.add_subplot(1, 3, 2)

ax2.hist(UEFA_data['xGA'])
plt.title('Expectativa de gols en contra')

ax3 = fig.add_subplot(1, 3, 3)

ax3.hist(UEFA_data['xGD'])
plt.title('Diferència esperada de gols')
```

Tal i com podem veure en els tres histogrames, les variables xG (Expectativa de gols per al club), xGA (Expectativa de gols en contra per al club) i xGD (Diferència esperada de gols (xG - xGA) tenen distribucions diferents.

Tot i així podem veure que en tots els casos les distribucions no són simètriques, sinó que estan més o menys esbiaixades cap a la dreta. Això vol dir que hi ha previsions que es desvien de la tendència general per incloure incloure un major nombre de gols a favor o en contra, o una major diferència de gols.
